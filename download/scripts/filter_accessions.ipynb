# FILTER ACCESSIONS
This notebook filters NCBI genome accessions based on a metadata file containing genome quality metrics and species names. 

#### Import modules

```python

import csv
import pandas as pd
```

#### Read file

```python
meta = pd.read_csv('metadata.tsv', delimiter="\t")  
meta.head()
```

#### Remove duplicate organisms

```python
meta_filtered = meta.drop_duplicates(subset='Organism Name')
meta_filtered.head()
```

#### Print to tsv

```python
meta_filtered.to_csv('filtered_metadata.tsv', sep="\t") 
```
