# Rename assemblies

#### Import libraries

```python
import pandas as pd
import os 
from pathlib import Path
from os import listdir
from os.path import isfile, join
```

```python
meta = pd.read_csv('filtered_metadata.tsv', delimiter="\t")  
#meta.head()
```

```python

folder = './assemblies'
files = [f for f in listdir(folder) if isfile(join(folder, f))]
sorted_files = sorted(files)
file_df = pd.DataFrame(sorted_files)
file_df.columns = ['filename']
#file_df.head()
```

```python
meta_sorted = meta.sort_values(by=['Assembly Accession'])
#meta_sorted.head()
```

```python
file_df.reset_index(drop=True, inplace=True)
meta_sorted.reset_index(drop=True, inplace=True)
combined_df = pd.concat([file_df, meta_sorted], axis=1)
#combined_df.head()
```

```python
dir = 'assemblies/'
files = [f for f in listdir(folder) if isfile(join(folder, f))]
files_dir = [dir + f for f in files]
files_dir_df = pd.DataFrame(files_dir)
files_dir_df.columns = ['Path']
files_dir_df_sorted = files_dir_df.sort_values(by='Path')
#files_dir_df_sorted

    
```

```python
files_dir_df_sorted.reset_index(drop=True, inplace=True)
combined_df_final = pd.concat([files_dir_df_sorted, combined_df], axis=1)
combined_df_final['directory'] = 'assemblies'
combined_df_final['extension'] = '.fasta'
#combined_df_final.head()
```

```python
combined_df_final['new_name'] = combined_df_final['Organism Name'] + combined_df_final['extension']
combined_df_final.tail()
```

```python
combined_df_final['new_name'] = combined_df_final['new_name'].replace(' ', '_', regex=True)
combined_df_final.tail()
```

```python
combined_df_final.columns = combined_df_final.columns.str.replace(" ", "_")
combined_df_final
```

```python
combined_df_final.to_csv('metadata_modified.tsv', sep="\t", index = False) 
```
